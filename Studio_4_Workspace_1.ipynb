```pyspark
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("PySpark Basics") \
    .getOrCreate()

data = [
    ("Alice", 34),
    ("Bob", 45),
    ("Cathy", 29),
    ("David", 34)
]

columns = ["name", "age"]

df = spark.createDataFrame(data, columns)
df.show()
```

```pyspark
df.printSchema()

spark.stop()
```
